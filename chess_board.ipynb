```python
import numpy as np
import matplotlib.pyplot as plt


class ChessBoard:
    def __init__(self):
        self.grid = np.zeros((8,8,3))
        self.red_col = 0
        self.blue_col = 0
        self.red_row = 0
        self.blue_row = 0
        
        for col in range(8):
           for row in range(8):
             if col %2==0 and row %2 ==0 or col %2==1 and row %2 ==1:
                self.grid[col][row] =  [1,1,1]
                
    def add_red(self,col,row):
       self.grid[col][row] =  [1,0,0]
       self.red_col = col
       self.red_row = row
    def add_blue(self,col,row):
       self.grid[col][row] =  [0,0,1] 
       self.blue_col = col
       self.blue_row = row 
       
    def render(self):
       plt.imshow(self.grid)
    def is_under_attack(self):
       attack = False
       if self.red_col == self.blue_col or self.red_row == self.blue_row :
          attack = True
       if abs(self.red_col - self.red_row)  == abs(self.blue_col - self.blue_row) :
          attack = True
       return attack     
         
                  

same_row= ChessBoard()
same_row.add_red(1,0)
same_row.add_blue(5,0)
same_row.render()
same_row.is_under_attack()
assert same_row.is_under_attack() == True
same_col= ChessBoard()
same_col.add_red(5,0)
same_col.add_blue(5,3)
same_col.render()
same_col.is_under_attack()
assert same_col.is_under_attack() == True
same_diagonal= ChessBoard()
same_diagonal.add_red(0,0)
same_diagonal.add_blue(2,2)
same_diagonal.render()
same_diagonal.is_under_attack()
assert same_diagonal.is_under_attack() == True
other_coordinates= ChessBoard()
other_coordinates.add_red(0,0)
other_coordinates.add_blue(4,2)
other_coordinates.render()
other_coordinates.is_under_attack()
assert other_coordinates.is_under_attack() == False
  
```
